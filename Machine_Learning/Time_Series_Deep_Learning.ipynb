```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
```

```python
import pandas as pd
df = pd.read_csv("../data/daily-total-female-births-in-cal.csv")
df.head()
```

```python
df['shifted'] = df['Daily total female births in California, 1959'].shift(1)
df = df[:-1]
df.tail()
```

```python
df['Month'] = [x.split('-')[1] for x in df['Date']]
```

```python
df2 = df.groupby('Month').mean()
df2.columns = ["Birth Month Average", "Drop"]
del df2['Drop']
df2['Month'] = df2.index
df2.head()
```

```python
pd.merge(df, df2, left_on='Month', right_on='Month')
```
